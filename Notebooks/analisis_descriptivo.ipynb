## üìÇ Cargar una sesi√≥n o evento

El objeto central es `fastf1.core.Session`.  
Para obtenerlo usamos `fastf1.get_session()`:

```python
import fastf1

session = fastf1.get_session(2021, 7, 'Q')  # 7ma carrera, sesi√≥n de clasificaci√≥n
print(session.name)  # 'Qualifying'
print(session.date)  # fecha y hora
```

Cada sesi√≥n pertenece a un evento (fin de semana de carrera o test).
Podemos acceder a la informaci√≥n del evento desde el objeto session.event.
```python
print(session.event['EventName'])  # 'French Grand Prix'
print(session.event['EventDate'])  # fecha de la carrera
```
Otra forma de acceder a los eventos es con 
`fastf1.get_event()`. que te permite buscarlo por nombre o n√∫mero de ronda u otras caracteristicas sin necesidad de poner el nombre correcto del GP.
```python
event = fastf1.get_event(2021, 7) # 7ma carrera del 2021
session = event.get_race()  # obtener datos de la carrera

'''si quiero usar en vez de numero de evento, puedo usar el nombre del GP o una caracteristica'''
event = fastf1.get_event(2021, 'French Grand Prix')
print(event['EventName'])  # 'French Grand Prix'

```
Si quiero ver un calendario de las carreras de un a√±o, te lo da en el formato de dataframe con el metodo .get_event_schedule("a√±o").
